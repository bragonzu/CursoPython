#La simulación de Monte Carlo


Generamos dos numeros aletatorios entre 0 
(x) y 1 (y)

Calcularemos x^2 + y^2

Si el valor es inferior a 1 -> estamos dentro del circulo

Si el valor es superior a 1 -> estamos fuera del circulo

Calcularemos el numero total de veces que estan dentro del circulo y lo dividimos entre el numero total de intentos para obtener dicha robabilidad para aproximar el valor de pi

Usamos  dicha probabilidad para aproximar el valoor de pi.

Repetimos el experimiento un numero suficiente de veces ( por ejm 1000), para obtener (1000) diferentes aproximaciones de pi.

Calculamos el promedio de los 1000 experimentos anteriores para dar un valor final de pi.

```python
def pi_montecarlo(n,n_exp):
  pi_avg=0
  pi_value_list=[]
  n=1000
  for i in range(100):
    value=0
    x=np.random.uniform(0,1,n).tolist()
    y=np.random.uniform(0,1,n).tolist()
    for j in range(n):
      z=np.sqrt(x[j]* x[j] + y[j] * y[j])
      if z<=1:
        value+=1
    float_value=float(value)
    pi_value=float_value*4 /n
    pi_value_list.append(pi_value)
    pi_avg+=pi_value
  pi=pi_avg/n_exp
  print(pi)
  fig=plt.plot(pi_value_list)
  return(pi,fig)
```

```python
pi_montecarlo(1000,200)
```
